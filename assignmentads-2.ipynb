# ii. Store all the metadata of the actors in the network:

```python
import networkx as nx

# Load the graph from the karate.gml file
G = nx.read_gml('karate.gml')

# Get the metadata attribute for each node
metadata_attribute = 'id'  # Replace 'label' with the actual attribute name from your GML file
metadata = nx.get_node_attributes(G, metadata_attribute)

# Print metadata for each node
for node, data in metadata.items():
    print(f'Node {node}: {metadata_attribute.capitalize()} {data}')
```

# iii. Calculate all types of centrality and analyze nodes:

```python
# Degree centrality
degree_centrality = nx.degree_centrality(G)

# Betweenness centrality
betweenness_centrality = nx.betweenness_centrality(G)

# Closeness centrality
closeness_centrality = nx.closeness_centrality(G)

# Eigenvector centrality
eigenvector_centrality = nx.eigenvector_centrality(G)

# PageRank centrality
pagerank_centrality = nx.pagerank(G)

# Print centrality measures for each node
for node in G.nodes():
    print(f'Node {node}:')
    print(f'Degree Centrality: {degree_centrality[node]}')
    print(f'Betweenness Centrality: {betweenness_centrality[node]}')
    print(f'Closeness Centrality: {closeness_centrality[node]}')
    print(f'Eigenvector Centrality: {eigenvector_centrality[node]}')
    print(f'PageRank Centrality: {pagerank_centrality[node]}')
    print()
```

```python
for node in sorted(degree_centrality, key=degree_centrality.get, reverse=True):
  print(node, degree_centrality[node])
```

```python
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos, node_color='pink')
nx.draw_networkx_edges(G, pos, width=0.5)
nx.draw_networkx_labels(G, pos, font_size=8)
nx.draw_networkx_edge_labels(G, pos, edge_labels={e: round(degree_centrality[e[0]], 2) for e in G.edges()}, label_pos=0.3)
plt.title('Degree Centrality')
plt.show()
```

# iv. Find possible k-components and clustering coefficient:

```python
# Find possible k-components
k_components = nx.k_components(G)

# Print possible k-components
print('Possible k-components:')
for k, components in k_components.items():
    print(f'k={k}: {components}')

# Clustering coefficient
clustering_coefficient = nx.average_clustering(G)
print(f'Clustering Coefficient: {clustering_coefficient}')
```

# v. Find communities using Girvan-Newman and Louvain algorithms:

```python
from networkx.algorithms import community

# Girvan-Newman algorithm
gn_communities = community.girvan_newman(G)

print('Communities from Girvan-Newman algorithm:')
for communities in next(gn_communities):
    print(communities)

# Louvain algorithm
louvain_communities = community.greedy_modularity_communities(G)

print('Communities from Louvain method:')
for i, communities in enumerate(louvain_communities, 1):
    print(f'Community {i}: {communities}')
```
